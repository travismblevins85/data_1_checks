```python
import requests
import pandas as pd
```

```python
# https://covidtracking.com/data/api
r = requests.get('https://api.covidtracking.com/v1/states/current.json')

data = r.json()
df = pd.DataFrame(data)
df.head()
```

```python
# Has multiple columns with all null values, so drop them for better visual
df.drop(['commercialScore', 'negativeRegularScore', 'positiveScore', 'score','dataQualityGrade', 'negativeScore', 'grade'], axis=1)
```
